```python
r = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
response = r.text
```

```python
r.text
```

```python
# 得到JSON格式的字串
```

```python
import json
```

```python
data = json.loads(response) #利用json解析網頁內容
```

```python
for d in data:
    print(d['title']) #取出每篇文章的標題
```

```python
for d in data:
    print(d['commentCount'])  # 留言數量
```

```python
for i in data:
    print(i['likeCount'])
```
